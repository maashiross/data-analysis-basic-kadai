```python
import numpy as np
import matplotlib.pyplot as plt

# データ準備
x = np.linspace(0, 10, 100)
y1 = x ** 2                      # 折れ線グラフ用
y2 = np.sin(x)                    # sin関数
data = np.random.randn(1000)      # ヒストグラム用
x_scatter = np.random.rand(50)    # 散布図用
y_scatter = np.random.rand(50)

# Figure(全体の描画領域)とAxes(個別の描画領域)を作成
fig, axes = plt.subplots(2, 3, figsize=(12, 6))  # 2行3列の6領域
fig.suptitle("Matplotlibで複数グラフを1つのキャンバスに表示", fontsize=16)

# 1. 折れ線グラフ
axes[0, 0].plot(x, y1, color='blue', label='y = x^2')
axes[0, 0].set_title("折れ線グラフ")
axes[0, 0].set_xlabel("x")
axes[0, 0].set_ylabel("y")
axes[0, 0].legend()

# 2. sin関数
axes[0, 1].plot(x, y2, color='red', label='y = sin(x)')
axes[0, 1].set_title("sin関数")
axes[0, 1].legend()

# 3. ヒストグラム
axes[0, 2].hist(data, bins=30, color='green', alpha=0.7)
axes[0, 2].set_title("ヒストグラム")

# 4. 散布図
axes[1, 0].scatter(x_scatter, y_scatter, color='purple', alpha=0.7)
axes[1, 0].set_title("散布図")
axes[1, 0].set_xlabel("X")
axes[1, 0].set_ylabel("Y")

# 5. cos関数
axes[1, 1].plot(x, np.cos(x), color='orange', label='y = cos(x)')
axes[1, 1].set_title("cos関数")
axes[1, 1].legend()

# 6. 棒グラフ
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 4]
axes[1, 2].bar(categories, values, color='skyblue')
axes[1, 2].set_title("棒グラフ")

# レイアウト調整
plt.tight_layout(rect=[0, 0, 1, 0.95])  # タイトルと重ならないように調整
plt.show()
```
