```python
!pip install japanize-matplotlib

import matplotlib.pyplot as plt
import japanize_matplotlib
import pandas as pd

# --- データ読み込み ---
df = pd.read_csv("sample_pandas_6.csv")
category_df = pd.read_csv("category.csv")  # ← 必要に応じて読み込み

# --- カテゴリー情報を結合 ---
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

# =========================================================
# ① カテゴリー列の出現頻度をカウントして棒グラフで表示
# =========================================================
category_counts = df['カテゴリー'].value_counts()

category_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('カテゴリー別 出現頻度')
plt.xlabel('カテゴリー')
plt.ylabel('出現回数')
plt.show()

# =========================================================
# ② 商品番号ごとに注文数の基本統計量を求める
# =========================================================
stats = df.groupby('商品番号')['注文数'].describe()
print("商品番号ごとの注文数 基本統計量")
print(stats)

# =========================================================
# 商品番号ごとにカテゴリー頻度をグラフ化
# =========================================================
for product_id, group in df.groupby('商品番号'):
    counts = group['カテゴリー'].value_counts()
    counts.plot(kind='bar', color='lightgreen', edgecolor='black')
    plt.title(f'商品番号 {product_id} のカテゴリー頻度')
    plt.xlabel('カテゴリー')
    plt.ylabel('出現回数')
    plt.show()
```
